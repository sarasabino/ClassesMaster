```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta_real_0=0
theta_real_1=1.3
m=250 #numero de muestras

xs=np.random.normal(20.0,10.0,m)

ys =xs * theta_real_1 +theta_real_0 +np.random.normal(0,5.0,m)

plt.scatter(xs,ys)
```

```python
def cost_function(theta_0,theta_1):
    
    count=0
    for i in range(0,m):
        count+= (xs[i] * theta_1 +theta_0-ys[i])**2
    
    function_cost=(1/m)*count
    return function_cost
cost_function(0.5,0.3)
```

```python
#pintar la hipotesis de la funcion de costes

theta_real_1h = 1
theta_real_0h = 0
ys_hipotesis = xs * theta_real_1h +theta_real_0h

plt.scatter(xs,ys,color='r')
plt.plot(xs,ys_hipotesis,color='b')
```

### Partial differentials of the cost function

$$\frac{\partial}{\partial\theta_0} = \frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i)$$

$$\frac{\partial}{\partial\theta_1} = \frac{1}{m}\sum_{i=1}^m(h_\theta(x_i)-y_i) \cdot x_i$$

```python
learning_rate = 0.0001
theta_1 = 1
theta_0 = 0
ideal = 3
n_iterations = 15000

for j in range (0,n_iterations):
        
    if cost_function(theta_0,theta_1) < ideal: 
        break
        
    for i in range (0,m):
        gradient_theta0 = (1/m) * ((xs[i] * theta_1 +theta_0) - ys[i])

        gradient_theta1 = (1/m) * ((xs[i] * theta_1 +theta_0) - ys[i]) * xs[i]
    
    theta_1 = theta_1 - gradient_theta1 * learning_rate
    theta_0 = theta_0 - gradient_theta0 * learning_rate
print(cost_function(theta_0,theta_1))
```

```python
ys_real = xs * theta_1 +theta_0

plt.scatter(xs,ys,color='r')
plt.plot(xs,ys_real,color='b')
plt.plot(xs,ys_hipotesis,color='g')
```
