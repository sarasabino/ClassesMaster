# NLTK

### Tokenizacin en frases

###### nltk.RegexpTokenizer

```python
import nltk
```

```python
t = nltk.RegexpTokenizer('\.', gaps=True) #patron y si lo que le doy es true o false
##es obligatorio mandar un patron
```

```python
t.tokenize(text)
```

###### gaps = True

```python
pattern = r'\.'
```

```python
pattern = r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<![A-Z]\.)(?<=\.|\?|\!)\s' #expresion regular de la documentacion de nltk
```

```python
t = nltk.RegexpTokenizer(pattern, gaps=True) 
```

```python
t.tokenize(text)
```

###### nltk.sent_tokenize

```python
#para tocanizar en frases diractamente un texto



nltk.sent_tokenize(text)
```

```python
quijote = open("./data/quijote.txt", encoding = "utf8").read()
quijote[:600]
```
